```python
xgb=XGBClassifier(n_estimators=100,learning_rate=0.1, eval_metric = 'auc')
ada=AdaBoostClassifier(DecisionTreeClassifier(max_depth=5),n_estimators=100,learning_rate=0.1)
knn=KNeighborsClassifier()
mlp=MLPClassifier()
lr=LogisticRegression()
rf=RandomForestClassifier()
dt=DecisionTreeClassifier(max_depth=3) 

models=[XGBClassifier(n_estimators=100,learning_rate=0.1, eval_metric = 'auc'),
        AdaBoostClassifier(DecisionTreeClassifier(max_depth=5),n_estimators=100,learning_rate=0.1),
        KNeighborsClassifier(),
        MLPClassifier(),
        DecisionTreeClassifier(max_depth=3)]
x_train_stack=np.zeros((x_train.shape[0],len(models)))
x_test_stack=np.zeros((x_test.shape[0],len(models)))
```

```python
n_folds=5
kf=KFold(n_splits=n_folds,shuffle=True)
for i , model in enumerate(models):
    x_stack_test_n=np.zeros((x_test.shape[0],n_folds))
    for j , (train_index,test_index) in enumerate(kf.split(x_train)):
        tr_x=x_train[train_index]
        tr_y=y_train[train_index]
        model.fit(tr_x,tr_y)
        
        x_train_stack[test_index,i]=model.predict(x_train[test_index])
        x_stack_test_n[:,j]=model.predict(x_test)
        
    x_test_stack[:,i]=x_stack_test_n.mean(axis=1)
```

```python
model_second=LogisticRegression()
model_second.fit(x_train_stack,y_train)
pred=model_second.predict(x_test_stack)
auc = roc_auc_score(y_test,pred)
acc = accuracy_score(y_test,pred)
pre = precision_score(y_test,pred)
rec = recall_score(y_test,pred)
f1 = f1_score(y_test,pred)
print("auc:%.3f"%auc)
print("准确率:%.3f%%"%(acc*100))
print ("查准率:%.3f%%"%(pre*100))
print ("召回率:%.3f%%"%(rec*100))
print ("F1:%.3f%%"%(f1*100))
```
